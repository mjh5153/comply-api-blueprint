```python
# write class

def transform_family_size(num):
  if num == 1:
    return 'alone'
  elif num > 1 and num < 5:
    return 'small'
  else:
    return 'large'
```

```python
df['family_size'].apply(transform_family_size) # Notes: apply() executes function on dataframe family_size column, passing column value to function parameter num
```

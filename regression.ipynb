```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('package(in lakhs)')
```

```python
df.head()
```

```python
X = df.iloc[:, 0:1]
X
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict([[2]])
```

```python
lr.predict([[9.5]])
```

```python
b = lr.intercept_
```

```python
m = lr.coef_
```

```python
print(m*9.5+b)
```

```python
#MAE : Mean absolute error : Average error in the prediction
# R2 score : How good my model is (Close to 1 -> good model , close to 0 -> bad model)
#MSE : penalise large error more

y_pred = lr.predict(X_test)
```

```python
#how successful your model is
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
#if the value is close to 1 then it is a successful model otherwise it is a failed model
```

```python
y_pred
```

```python
#avg error in the prediction
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, y_pred))
```

```python
#mean squared error (MSE)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
